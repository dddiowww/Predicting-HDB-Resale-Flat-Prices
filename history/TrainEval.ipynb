# Train & Eval

```python
# ==========================================================
# CV + Hyperparam Search with tqdm, export CSV submissions
# Models: Linear Regression (Ridge) / RandomForest / XGBoost / LightGBM
# ==========================================================
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn.model_selection import KFold, ParameterGrid
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor


# ========== Data Preparation ==========
TARGET_COL = "RESALE_PRICE"
features = [c for c in df_train.columns if c != TARGET_COL]
print(f"Number of features: {len(features)}")
X = df_train[features].values
y = df_train[TARGET_COL].values
X_test = df_test[features].values


# ========== Utility Functions ==========
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))


def cv_search(model_name, model_ctor, param_grid, X, y, n_splits=5, random_state=42):
    """K-Fold cross-validation grid search"""
    best_params, best_rmse = None, float("inf")
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=random_state)
    grid = list(ParameterGrid(param_grid))
    pbar = tqdm(grid, desc=f"{model_name} grid", leave=True)

    for params in pbar:
        fold_rmses = []
        for tr_idx, va_idx in kf.split(X):
            X_tr, X_va = X[tr_idx], X[va_idx]
            y_tr, y_va = y[tr_idx], y[va_idx]

            model = model_ctor(**params)
            model.fit(X_tr, y_tr)
            pred = model.predict(X_va)

            rmse_real = rmse(np.exp(y_va), np.exp(pred))
            fold_rmses.append(rmse_real)

        mean_rmse = float(np.mean(fold_rmses))
        if mean_rmse < best_rmse:
            best_rmse, best_params = mean_rmse, params
            print(f"\n New best for {model_name}: RMSE={best_rmse:.0f}")
            for k, v in best_params.items():
                print(f"  {k}: {v}")
            print()

        pbar.set_postfix({
            "cv_rmse": f"{mean_rmse:.0f}",
            "best": f"{best_rmse:.0f}"
        })

    print(f"\n===== {model_name} Best Result =====")
    print(f"CV RMSE (real price): {best_rmse:.0f}")
    print("Best Params:")
    for k, v in best_params.items():
        print(f"  {k}: {v}")
    print("=" * 40 + "\n")

    return best_params, best_rmse


def train_full_and_predict(model_ctor, params, X, y, X_test):
    model = model_ctor(**params)
    model.fit(X, y)
    pred_log = model.predict(X_test)
    pred_price = np.exp(pred_log)
    return pred_price


def save_submission(preds, path):
    sub = pd.DataFrame({"Id": np.arange(len(preds)), "Predicted": preds})
    sub.to_csv(path, index=False)
    print(f"Saved: {path}")
    return sub


# ========== Parameter Grids ==========
ridge_grid = {
    "alpha": [0.0, 0.1, 1.0, 10.0],
    "random_state": [42]
}

rf_grid = {
    "n_estimators": [300, 500],
    "max_depth": [10, 15],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2],
    "random_state": [42],
    "n_jobs": [-1]
}

xgb_grid = {
    "n_estimators": [1300, 1600],
    "learning_rate": [0.05],
    "max_depth": [8, 10],
    "subsample": [0.8],
    "colsample_bytree": [0.8],
    "reg_lambda": [2.0],
    "random_state": [42],
    "n_jobs": [-1],
    "tree_method": ["hist"],
    "device": ["cuda"]
}

lgbm_grid = {
    "n_estimators": [1500, 2000],
    "learning_rate": [0.05],
    "max_depth": [-1, 8],
    "subsample": [0.8],
    "colsample_bytree": [0.8],
    "reg_lambda": [1.0, 3.0],
    "reg_alpha": [0, 1],
    "random_state": [42],
    "n_jobs": [-1],
    "device": ["gpu"]
}


# ========== Run CV Search ==========
# ridge_best_params, ridge_cv = cv_search("LinearRidge", Ridge, ridge_grid, X, y)
# rf_best_params, rf_cv = cv_search("RandomForest", RandomForestRegressor, rf_grid, X, y)
xgb_best_params, xgb_cv = cv_search("XGBoost", XGBRegressor, xgb_grid, X, y)
lgbm_best_params, lgbm_cv = cv_search("LightGBM", LGBMRegressor, lgbm_grid, X, y)


# ========== Train Full + Predict + Save ==========
# ridge_preds = train_full_and_predict(Ridge, ridge_best_params, X, y, X_test)
# rf_preds = train_full_and_predict(RandomForestRegressor, rf_best_params, X, y, X_test)
xgb_preds = train_full_and_predict(XGBRegressor, xgb_best_params, X, y, X_test)
lgbm_preds = train_full_and_predict(LGBMRegressor, lgbm_best_params, X, y, X_test)

# save_submission(ridge_preds, "linear_submission.csv")
# save_submission(rf_preds, "rf_submission.csv")
save_submission(xgb_preds, "xgb_submission.csv")
save_submission(lgbm_preds, "lgbm_submission.csv")


# ========== Compare CV Scores ==========
cv_scores = {
    # "LinearRidge": ridge_cv,
    # "RandomForest": rf_cv,
    "XGBoost": xgb_cv,
    "LightGBM": lgbm_cv
}

best_name = min(cv_scores, key=cv_scores.get)
print(f"\n Best by CV RMSE: {best_name} ({cv_scores[best_name]:.2f})")

best_preds = {
    # "LinearRidge": ridge_preds,
    # "RandomForest": rf_preds,
    "XGBoost": xgb_preds,
    "LightGBM": lgbm_preds
}[best_name]
save_submission(best_preds, "submission_best.csv")

print("\nFiles saved:")
# print(" - linear_submission.csv")
# print(" - rf_submission.csv")
print(" - xgb_submission.csv")
print(" - lgbm_submission.csv")
print(" - submission_best.csv (best by CV)")
```
